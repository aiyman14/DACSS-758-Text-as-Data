# Regex

```python
import re
```

```python
test_string = "Today's date is 2-22-26"

print(re.findall(r"[\d-]+", test_string))
print(re.findall(r"[2-22-26]", test_string))
print(re.findall(r"[\\d-]+", test_string))
print(re.findall(r"[0-9]+", test_string))
```

```python
tester_string = "For 3 years I lived at 127 Lincoln Rd. The 1st year was great."
```

```python
re.search("Lincoln", tester_string)
```

```python
re.findall("Lincoln", tester_string)
```

```python
re.findall("road", tester_string)
```

```python
# disjuctions
print(re.findall("for", tester_string))
print(re.findall("[Ff]or", tester_string))
print(re.findall("[0123456789]", tester_string))
```

```python
# ranges
print(re.findall("[0-9]", tester_string))
print(re.findall("[A-Z]", tester_string))
print(re.findall("[A-Za-z]", tester_string))
```

```python
# negations
print(re.findall("[^A-Za-z]", tester_string))
print(re.findall("[^A-Za-z\\s]", tester_string))
```

```python
# optional characters and counters
print(re.findall("years?", tester_string))
print(re.findall("lived?", tester_string))
print(re.findall("[0-9]+", tester_string))
print(re.findall("[0-9]{3}", tester_string))
```

```python
# wild cards
print(re.findall(".", tester_string))
print(re.findall("\\.", tester_string))  # escaping
print(re.findall(". years", tester_string))
```

```python
# anchors
print(re.findall("^.+years", tester_string))
print(re.findall("years.+$", tester_string))
```

```python
# more operators
print(re.findall("\\d", tester_string))  # digits
print(re.findall("\\D", tester_string))  # non-digits
print(re.findall("\\w", tester_string))  # alpha numeric
print(re.findall("\\W", tester_string))  # non alpha numeric
print(re.findall("\\s", tester_string))  # white space
print(re.findall("\\S", tester_string))  # non white space
```

```python
# find and replace
print(re.sub("Rd", "Road", tester_string))
print(re.sub("year(s)?", "month\\1", tester_string))
```

```python
common_nouns = [token.text.lower() for token in doc if token.pos_ == "NOUN" and not token.is_stop and not token.is_punct]
noun_freq = Counter(common_nouns)
most_common_nouns = noun_freq.most_common(10)
most_common_nouns
```

```python
# named entity recognition (NER)
cols=['text','start_char','end_char','ent.label']
rows = []
for ent in doc.ents:
    row = [ent.text, ent.start_char, ent.end_char, ent.label_]
    rows.append(row)
```
